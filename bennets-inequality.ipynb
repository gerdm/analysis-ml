# Bennet's inequality

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
```

## Understanding Equation 2.13

The claim is that
$$
    \sum_{k=0}^\infty \frac{(\lambda b)^k}{(k+2)!} \leq \frac{1}{2}\sum_{k=0}^\infty \frac{(\lambda b)^k}{3^k}.
$$

We then show
$$
    \frac{1}{(k+2)!} \leq \frac{1}{2\cdot3^k} \,\forall k\in\mathbb{Z}^+.
$$

```python
xdom = np.arange(0, 10)
```

```python
exp_term = (xdom + 1) * (xdom + 2) * factorial(xdom)
exp_term = 1 / exp_term

ubound_term = 1 / (2 * 3 ** xdom)
```

```python
for x, t, ubt in zip(xdom, exp_term, ubound_term):
    print(f"{x=} | {t=:0.3e} | {ubt=:0.3e}")
```

```python
plt.plot(xdom, exp_term, label="exp-term", marker="o")
plt.plot(xdom, ubound_term, label="ubound", marker="o")
plt.legend()
plt.yscale("log")
plt.grid(alpha=0.3)
```
